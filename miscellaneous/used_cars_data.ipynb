# Used cars dataset — exploratory data analysis

Structure, dimensions, missingness, and per-column summaries (**`describe()`** for numeric fields, **`value_counts()`** for categorical text). Run this notebook with a **Python** kernel (e.g. Anaconda base interpreter).

## Imports and file path

Libraries and path to `used_cars_data.csv` (same folder as this notebook).

```python
# =============================================================================
# Imports — libraries and path to the CSV (same folder as this notebook)
# =============================================================================
from pathlib import Path

import numpy as np
import pandas as pd
from IPython.display import Markdown, display

DATA_PATH = Path("used_cars_data.csv")
```

## Load data from CSV

Read the full table into a pandas `DataFrame`. Running the next cell also writes `used_cars_head_preview.pdf` and `used_cars_head_preview.jpg` in this folder (static table image of `df.head()` for reports or slides).

```python
# =============================================================================
# Load CSV — one row per listing; column names come from the first row
# Also write df.head() to PDF + JPG for easy copy/paste into slides or reports
# =============================================================================
import matplotlib.pyplot as plt

df = pd.read_csv(DATA_PATH)
print(f"Loaded {len(df):,} rows, {df.shape[1]} columns.")

head_preview = df.head()
display(head_preview)

HEAD_EXPORT = Path("used_cars_head_preview")


def save_head_table_pdf_jpg(frame: pd.DataFrame, path_base: Path) -> None:
    """Render `frame` as a simple table and save as .pdf and .jpg (same layout)."""
    nrows, ncols = frame.shape
    fig_w = max(16.0, 0.55 * ncols)
    fig_h = max(3.0, 0.55 * (nrows + 1))
    fig, ax = plt.subplots(figsize=(fig_w, fig_h))
    ax.axis("off")
    cell_text = frame.astype(str).values.tolist()
    col_labels = [str(c) for c in frame.columns]
    row_labels = [str(i) for i in frame.index]
    tbl = ax.table(
        cellText=cell_text,
        colLabels=col_labels,
        rowLabels=row_labels,
        loc="center",
        cellLoc="left",
    )
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(7)
    tbl.scale(1.0, 1.6)
    fig.tight_layout()
    pdf_path = path_base.with_suffix(".pdf")
    jpg_path = path_base.with_suffix(".jpg")
    fig.savefig(pdf_path, bbox_inches="tight", dpi=200)
    fig.savefig(jpg_path, bbox_inches="tight", dpi=200)
    plt.close(fig)
    print(f"Saved table snapshot: {pdf_path.resolve()}")
    print(f"Saved table snapshot: {jpg_path.resolve()}")


save_head_table_pdf_jpg(head_preview, HEAD_EXPORT)
```

## Dataset overview (aggregate only)

Column names, row/column counts, dtype of each column, and **missing value counts** per column. No per-column frequency or distribution detail here—only table-level structure.

The next cell also saves **`used_cars_dataset_overview.pdf`** and **`used_cars_dataset_overview.jpg`** (single page: dimensions line + full table) for pasting into Google Docs or slides.

```python
# =============================================================================
# Overview — headers, dimensions, dtypes, missing counts (no value breakdowns)
# Also writes PDF + JPG snapshots for Google Docs / slides (same folder as notebook)
# =============================================================================
import matplotlib.pyplot as plt

n_rows, n_cols = df.shape
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(2)

display(
    Markdown(
        "### Dataset overview\n\n"
        f"| | **Value** |\n"
        f"| --- | ---: |\n"
        f"| **Rows** | {n_rows:,} |\n"
        f"| **Columns** | {n_cols} |\n"
        f"| **Total missing cells** | {missing.sum():,} |\n"
    )
)

summary = pd.DataFrame(
    {
        "#": range(1, n_cols + 1),
        "column": df.columns.astype(str),
        "dtype": df.dtypes.astype(str).values,
        "missing_count": missing.values,
        "missing_pct": missing_pct.map(lambda x: f"{x:.2f}%").values,
    }
).set_index("#")

display(Markdown("**Columns (order), dtypes, and missing values**"))
display(summary)


def save_overview_table_pdf_jpg(frame: pd.DataFrame, path_base: Path, subtitle: str) -> None:
    """Render overview + full table; save as .pdf and .jpg for pasting into docs."""
    tbl = frame.reset_index()
    nrows, ncols = tbl.shape
    fig_w = max(18.0, 0.55 * ncols)
    fig_h = max(5.0, 0.42 * (nrows + 3))
    fig = plt.figure(figsize=(fig_w, fig_h))
    fig.suptitle("Used cars dataset — structure overview", fontsize=13, fontweight="bold", y=0.98)
    fig.text(0.5, 0.93, subtitle, ha="center", fontsize=10)
    ax = fig.add_axes([0.04, 0.06, 0.92, 0.80])
    ax.axis("off")
    cell_text = tbl.astype(str).values.tolist()
    col_labels = [str(c) for c in tbl.columns]
    t = ax.table(
        cellText=cell_text,
        colLabels=col_labels,
        loc="center",
        cellLoc="left",
    )
    t.auto_set_font_size(False)
    t.set_fontsize(8)
    t.scale(1.0, 1.5)
    pdf_path = path_base.with_suffix(".pdf")
    jpg_path = path_base.with_suffix(".jpg")
    fig.savefig(pdf_path, bbox_inches="tight", dpi=200)
    fig.savefig(jpg_path, bbox_inches="tight", dpi=200)
    plt.close(fig)
    print(f"Saved overview snapshot: {pdf_path.resolve()}")
    print(f"Saved overview snapshot: {jpg_path.resolve()}")


overview_subtitle = (
    f"Rows: {n_rows:,}   |   Columns: {n_cols}   |   Total missing cells: {missing.sum():,}"
)
save_overview_table_pdf_jpg(summary, Path("used_cars_dataset_overview"), overview_subtitle)
```

## Per-column analysis

- **Numeric:** `model_year` is read as a number; `price` and `milage` are parsed from their string formats into `price_usd` and `milage_mi` for `describe()`.
- **Categorical (text/object):** `value_counts()` for each original object column (including raw `price` / `milage` strings if present). High-cardinality columns show all levels (may be long).

```python
# =============================================================================
# Parse list price and mileage strings for numeric descriptive statistics
# =============================================================================

def parse_price(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float)) and not isinstance(x, bool):
        return float(x)
    s = str(x).replace("$", "").replace(",", "").strip()
    if s == "" or s.lower() in {"nan", "none"}:
        return np.nan
    return float(s)


def parse_milage(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float)) and not isinstance(x, bool):
        return float(x)
    s = str(x).lower().replace("mi", "").replace(",", "").strip()
    if s == "" or s in {"–", "-", "nan"}:
        return np.nan
    try:
        return float(s)
    except ValueError:
        return np.nan


df_eda = df.copy()
df_eda["price_usd"] = df_eda["price"].map(parse_price)
df_eda["milage_mi"] = df_eda["milage"].map(parse_milage)

numeric_cols = [c for c in ["model_year", "price_usd", "milage_mi"] if c in df_eda.columns]
print("Numeric columns used for describe():", numeric_cols)
df_eda[numeric_cols].describe().T
```

```python
# =============================================================================
# Categorical columns — value_counts() per object-dtype column
# =============================================================================
cat_cols = df.select_dtypes(include=["object", "string"]).columns.tolist()
print("Object (categorical) columns:", cat_cols)
print()

for col in cat_cols:
    print("=" * 72)
    print(f"{col}  |  dtype={df[col].dtype}  |  n_unique={df[col].nunique(dropna=False)}")
    print(df[col].value_counts(dropna=False).to_string())
    print()
```
